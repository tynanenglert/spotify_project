```python
# List of files to merge 
file_paths = [
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2012-2017_0.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2017-2018_1.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2018-2019_2.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2019_3.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2019-2020_4.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2020_5.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2020-2021_6.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2021-2022_7.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2022_8.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2022-2023_9.json',
    '/Users/brookecoffey/Desktop/DataAnalyticsTy/Achievement 6 /Spotify Project/01 Data/01 Original Data/Streaming_History_Audio_2023_10.json'
]

# Create dictionary to store dataframes
dataframes = {}

# Loop through each file and read it into a dataframe
for file_path in file_paths:
    df = pd.read_json(file_path)
    file_name = file_path.split('/')[-1]  # Extract the file name from the path
    dataframes[file_name] = df

# Create a dictionary to store column data types
column_data_types = {}

# Loop through each DataFrame and extract column data types
for file_name, df in dataframes.items():
    data_types = df.dtypes
    column_data_types[file_name] = data_types

# Create a DataFrame from the column data types dictionary
data_types_df = pd.DataFrame(column_data_types)

# Display the data types DataFrame
print(data_types_df)
```

### Streaming_History_Audio_2012-2017_0.json \ has episode_name, episode_show_name, and spotify_episode_uri as floats, but should be objects 
### Streaming_History_Audio_2023_10.json \ has skipped as boolean, but should be float
### Streaming_History_Video_2017-2023.json has offline and offline_timestamp as floats, but should be bool and int64 repectively

```python
# Setting column data types in merged dataframe 
merged_df = pd.DataFrame()
```

```python
# Setting a dictionary of column names and data types 
column_data_type = {
    'skipped':'object', 
    'offline':'object',
    'offline_timestamp':'int64'
}
```

```python
# Defining a function to set column data types while merging the datasets
for file_path in file_paths:
    df = pd.read_json(file_path)
    
    # Setting data types for specific columns
    for column, dtype in column_data_types.items():
        if column in df.columns:
            df[column] = df[column].astype(dtype)
    
    # Concatenate the DataFrame to the merged_df
    merged_df = pd.concat([merged_df, df], ignore_index=True)
```
